```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
chicago = pd.read_csv("chicago.csv").dropna(how='all')
chicago.tail()
```

```python
chicago.info()
```

```python
chicago.nunique()
```

임금 데이터를 숫자 값으로 변환 작업

```python
chicago['Employee Annual Salary']=chicago["Employee Annual Salary"].str.replace("$",'').astype(float)
```

```python
import warnings
warnings.filterwarnings('ignore')
```

boxplot() 함수 이용한 데이터 시각화 (부문별 임금 데이터)

```python
plt.figure(figsize=(16,4))
sns.boxplot(data=chicago, x='Department', y='Employee Annual Salary')
plt.show()
```
